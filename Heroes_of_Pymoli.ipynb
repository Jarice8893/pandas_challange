```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
heroes_data = pd.read_csv("Resources/purchase_data.csv")
heroes_data.head()
```

```python
total_players = len(heroes_data["SN"].unique())
total_players

total_players_count = pd.DataFrame({"Total Players": [total_players]})
total_players_count
```

```python
unique_items_count = len(heroes_data["Item ID"].unique())
average_purchase = heroes_data["Price"].mean()
average_purchase_count = round(average_purchase,2)
total_purchase_count = heroes_data["Purchase ID"].count()
total_revenue_count = average_purchase_count*total_purchase_count

purchasing_analysis = pd.DataFrame({'Number of Unique Items': [unique_items_count], 'Average Price': f'${average_purchase_count}', 'Number of Purchases': (total_purchase_count), 'Total Revenues': f'${total_revenue_count}'})
purchasing_analysis
```

```python
genders = heroes_data.groupby("Gender")

bygender_players = genders.nunique()["SN"]
#bygender_players = pd.series.round
#bygender_players = round(bygender_players,2)
#bygender_players = pd.DataFrame.style.set_precision(2)
total_gender_percent = bygender_players/total_players * 100
gender_analysis = pd.DataFrame({"Total Count": bygender_players, "Percentage Of Players":total_gender_percent})
#gender_analysis = style.format({"Percentage Of Players":'{:,.2f}'.format})
#gender_analysis = pd.Series({"Percentage of Players"}:"{:,.2f}")
gender_analysis
```

```python
purchase_count = genders["Purchase ID"].count()
avg_purchase_price = genders["Price"].mean()
total_purchase_value = genders["Price"].sum()
avg_total_purchase_per_person = total_purchase_value/bygender_players
gender_analysis = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price":avg_purchase_price, "Average Purchase Value":total_purchase_value, "Avg Purchase Total per Person":avg_total_purchase_per_person})
gender_analysis
```

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
heroes_data["Age Group"] = pd.cut(heroes_data["Age"],age_bins, labels=group_names)
heroes_data

age = heroes_data.groupby("Age Group")

total_count_age = age.nunique()["SN"]

percentage_players = (total_count_age/total_players)* 100

age_analysis = pd.DataFrame({"Total Count": total_count_age, "Percentage Of Players":percentage_players})

age_analysis
```

```python
purchase_count_age = age["Purchase ID"].count()
avg_purchase_price_age = age["Price"].mean()
total_purchase_value_age = age["Price"].sum()
avg_total_purchase_per_person_age = total_purchase_value_age/total_count_age
age_analysis = pd.DataFrame({"Purchase Count": purchase_count_age, "Average Purchase Price":avg_purchase_price_age, "Average Purchase Value":total_purchase_value_age, "Avg Purchase Total per Person":avg_total_purchase_per_person_age})
age_analysis
```

```python
spenders = heroes_data.groupby("SN")
purchase_count_spenders = spenders["Purchase ID"].count()
avg_purchase_price_spenders = spenders["Price"].mean()
total_purchase_value_spenders = spenders["Price"].sum()

top_spenders = pd.DataFrame({"Purchase Count": purchase_count_spenders, "Average Purchase Price":avg_purchase_price_spenders, "Total Purchase Value":total_purchase_value_spenders})
formatted_spenders = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()
formatted_spenders
```

```python
items = heroes_data[["Item ID", "Item Name", "Price"]]
popular_items = items.groupby(["Item ID", "Item Name"])
purchase_count_item = popular_items["Price"].count()
purchase_value_item = (popular_items["Price"].sum())
item_price = purchase_value_item/purchase_count_item

most_popular_items = pd.DataFrame({"Purchase Count": purchase_count_item, "Item Price":item_price, "Total Purchase Value":purchase_value_item})
formatted_items = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()
formatted_items
```
